## Clone workshop repository from github

This is just for your own documentation since you had to do this to get the repository in the first place! 

*NOTE: You need to run this from the root of your Jupyter space. If you do it here you'll just make another copy of the repository inside the repository folder.*

Go to [my JupyterPythonIntro github repository](https://github.com/emonson/JupyterPythonIntro) and click on the green button that says "Clone or download". Copy the "Clone with HTTPS" URL and use the `git clone` command with a `bash` kernel to download the repository into your Jupyter space.

```bash
git clone https://github.com/emonson/JupyterPythonIntro.git
```
